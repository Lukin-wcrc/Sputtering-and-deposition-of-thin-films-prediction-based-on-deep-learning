# 데이터 임포트

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# 스케일 처리
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import  MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
def cor_sc(X,Y):
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(X)
    X_scaled = pd.DataFrame(X_scaled)
    
    cor = []
    for i in range(X.shape[1]):
        a = pd.DataFrame({'X':X_scaled.iloc[:,i], 'Y':Y.iloc[:,i]})
        b = a.corr()
        cor.append(b.iloc[1,0])
    cor = pd.DataFrame(cor)

    plt.figure(figsize=(6,8))
    sns.heatmap(cor, center=0)
    plt.xlabel('Correation')
    plt.ylabel('No.')
    plt.show()
```

```python
def cor(X,Y):
    cor = []
    for i in range(X.shape[1]):
        a = pd.DataFrame({'X':X.iloc[:,i], 'Y':Y.iloc[:,i]})
        b = a.corr()
        cor.append(b.iloc[1,0])
    cor = pd.DataFrame(cor)

    plt.figure(figsize=(6,8))
    sns.heatmap(cor, center=0)
    plt.xlabel('Correation')
    plt.ylabel('No.')
    plt.show()
    return cor
```

# 물질별 상관관계 차이 확인

## 독립이표본 t검정


```python
from scipy.stats import levene, ttest_ind
print(ttest_ind(cor_Al,cor_Ti))
# print(ttest_ind(V3,B7))
```

## 비모수적 독립표본 t 검정  (Mann-Whitney U Test / Wilcoxon rank sum test)

```python
# 정규성 확인 
from  scipy.stats import shapiro
print(shapiro(cor_M['MG'][cor_M['M']==0]))
print(shapiro(cor_M['MG'][cor_M['M']==1] ))
```

p-value가 0.05보다 크므로 각 집단의 자료가 정규성을 만족한다고 볼 수 없다.

```python
#등분산성 확인 - 레빈 검증
from scipy.stats import levene
print(levene(cor_M['MG'][cor_M['M']==0],
      cor_M['MG'][cor_M['M']==1]))

#등분산성 확인 - 바틀렛 검증
from scipy.stats import bartlett
print(bartlett(cor_M['MG'][cor_M['M']==0],
      cor_M['MG'][cor_M['M']==1]))
```

```python
# 일원분산분석
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

formula = 'MG ~ C(M)'
lm = ols(formula, cor_M).fit()
print(anova_lm(lm))
```

```python
from scipy.stats import mannwhitneyu, ranksums
print(mannwhitneyu(cor_Al, cor_Ti))
print(ranksums(cor_Al, cor_Ti))
```

두 집단의 모분산 차이가 있다. 등분산성을 가정할 수 없다.
