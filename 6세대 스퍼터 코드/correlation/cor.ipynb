```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# 스케일 처리
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import  MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
def cor_sc(X,Y):
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(X)
    X_scaled = pd.DataFrame(X_scaled)
    
    cor = []
    for i in range(X.shape[1]):
        a = pd.DataFrame({'X':X_scaled.iloc[:,i], 'Y':Y.iloc[:,i]})
        b = a.corr()
        cor.append(b.iloc[1,0])
    cor = pd.DataFrame(cor)

    plt.figure(figsize=(6,8))
    sns.heatmap(cor, center=0)
    plt.xlabel('Correation')
    plt.ylabel('No.')
    plt.show()
```

```python
def cor(X,Y):
    cor = []
    for i in range(X.shape[1]):
        a = pd.DataFrame({'X':X.iloc[:,i], 'Y':Y.iloc[:,i]})
        b = a.corr()
        cor.append(b.iloc[1,0])
    cor = pd.DataFrame(cor)

    plt.figure(figsize=(6,8))
    sns.heatmap(cor, center=0)
    plt.xlabel('Correation')
    plt.ylabel('No.')
    plt.show()
```

```python
B7_TM_re2 = pd.read_csv('data/B7_TM_re2.csv')
B7_MG = pd.read_csv('data/B7_MG.csv')
V3_TM_re2 = pd.read_csv('data/V3_TM_re2.csv')
V3_MG = pd.read_csv('data/V3_MG.csv')
```

```python
plt.figure(figsize=(18,6))
B7_TM_re2.boxplot()
plt.show()
```

```python
plt.figure(figsize=(18,6))
V3_TM_re2.boxplot()
plt.show()
```

```python
cor_sc(B7_TM_re2,B7_MG)
```

```python
cor(B7_TM_re2,B7_MG)
```

```python
cor_sc(V3_TM_re2,V3_MG)
```

```python
cor(V3_TM_re2,V3_MG)
```
